For my web source I chose box office data of different movies from  Box Office Mojo. The tables I am working with are broken up by year and display the top 200 films from the box office each year.
```python
# Importing necessary packages
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
pd.set_option('display.max_columns', None)
```

#### Step 1: Scraping the data and merging each year's tables on a common key

Since I will need to scrape multiple tables from my web source, I will define a function that will scrape the table from the html of my source that I can pass each individual url into:

```python
# Defining a function to scrape a table off of my web source using beautiful soup since multiple tables will need to be scraped
def scrape_table(url):
    response = requests.get(url)
    
    # Checking if the response code is good
    if response.status_code == 200:
        html_content = response.text
        
        # Parsing the HTML
        soup = BeautifulSoup(response.content, 'html.parser')
        
        table = soup.find('table')
        df = pd.read_html(str(table), header=0)[0]
        return df
    
    else: print("Failed to retrieve the webpage.")
```

##### 2021 Data

```python
url_total_grosses_2021 = "https://www.boxofficemojo.com/year/2021/?grossesOption=totalGrosses"
url_calendar_grosses_2021 = "https://www.boxofficemojo.com/year/2021/?grossesOption=calendarGrosses"

total_grosses_2021 = scrape_table(url_total_grosses_2021)
calendar_grosses_2021 = scrape_table(url_calendar_grosses_2021)

# Merge data frames on the common key of film title ("Release")
# Keeping all data from each table for now and distinguishing columns with different suffixes
merged_2021 = pd.merge(total_grosses_2021, calendar_grosses_2021, on="Release", suffixes=('_total', '_calendar'))

# Display the merged DataFrame
display(merged_2021)
```

#### Step 3: Changing data type of release dates and closing dates to datetime and ensuring year is specified

```python
# Checking the data types of every column (should be the same for every year)
merged_2021.dtypes
```

I am going to want to combine all of the data from different years that I scraped into a singular data frame, however I need to ensure that I will know which films came out in what years. Luckily, there are already columns for opening/release dates and closing dates (if available) of each film. I checked the data types of every column above and can see that at the moment both the open and close columns are object types, and at the moment they only contain the month and the day that the film opened. I will need to convert both of these columns to datetime and enter they year, however there is a condition I need to account for in the closing date - if the month of the closing date falls before the month of the opening/release date, the closing date needs to be recorded as the following year of the opening/release date. Since I will need to convert all four of my current data frames, I will first create a function to convert these columns to datetime that takes this condition into account:

```python
"""Defining a function to convert open and close columns to datetime, with a condition that if the month of close falls 
before the month of open the a year would be added to the release year"""
def adjust_year(opening_date, closing_date, year):
    # Returning NaN if there is no closing date
    if closing_date is None:
        return None
    # Extracting month from the opening date from a "month abbreviated day" string
    opening_month = pd.to_datetime(opening_date, format='%b %d').month
    # Extracting month from the closing date from a "month abbreviated day" string
    closing_month = pd.to_datetime(closing_date, format='%b %d').month
    # If the closing month falls before the opening month, increase the specified year by 1
    if closing_month < opening_month:
        return pd.to_datetime(closing_date + ' ' + str(year + 1), format='%b %d %Y')
    # If that is not the case, keep the specified year
    else:
        return pd.to_datetime(closing_date + ' ' + str(year), format='%b %d %Y')
```

```python
# Using a lambda function to apply the adjust_year function to every row of the data frame
merged_2021['Close'] = merged_2021.apply(lambda row: adjust_year(row['Open'], row['Close'], 2021), axis=1)

# Converting the 'Open' and 'Close' date columns to datetime, appending all in the "Open" column with the year 2021
merged_2021['Open'] = pd.to_datetime(merged_2021['Open'] + ' 2021')
merged_2021['Close'] = pd.to_datetime(merged_2021['Close'])

# Displaying the data frame to check datetime conversions
display(merged_2021)
```

Now I can take this framework and apply it to the remaining years' data frames:

```python
merged_2020['Close'] = merged_2020.apply(lambda row: adjust_year(row['Open'], row['Close'], 2020), axis=1)

merged_2020['Open'] = pd.to_datetime(merged_2020['Open'] + ' 2020')
merged_2020['Close'] = pd.to_datetime(merged_2020['Close'])

display(merged_2020)
```

#### Step 8: Handling Re-releases

While scraping the tables from the web and working with the data frame, I noticed that a few of the films are re-releases. While on the web table this was denoted below the title, it seems that when scraped the year and the words "Re-Release" were tacked onto the title string. This could pose an issue when it comes to matching up films by title from all of my different data sources since having "Re-release" in the title would not make the films an exact match. To handle this I am going to coplete two major steps: first, I am going to create a seperate boolean column noting a film as a re-release based on the presence of the string "Re-release" in the title column. Once this data is preserved, I am going to use a regular expression to extract just the title of the film away from the year and re-release and keep just the title in the title column.

```python
# Creating a boolean column 'rerelease' that populated True if the title column contains the string "Re-release"
boxoffice_df['rerelease'] = boxoffice_df['title'].str.contains('Re-release', case=False)
boxoffice_df
```

```python
# Looking at all rows where 'rerelease'=True to determine what regular expression to use
boxoffice_df[boxoffice_df['rerelease']]
```

```python
# Defining a regular expression pattern to capture the title without the re-release info
regex_pattern = r'^(.*?)(?:\d{4} Re-release|10th Anniversary Re-release)?$'

# Extracting the title without the re-release info using str.extract()
boxoffice_df['title'] = boxoffice_df['title'].str.extract(regex_pattern, expand=False).str.strip()

# Checking that the extraction worked correctly
boxoffice_df[boxoffice_df['rerelease']]
```
