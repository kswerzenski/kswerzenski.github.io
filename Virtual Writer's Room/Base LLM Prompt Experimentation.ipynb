### Prompt 1: Character Design for a Specific Audience

This first prompt takes one of the non-copyrighted scripts from StageMilk entitled "Pictures of My Dog" and will ask the AI to review the scene from the perspective of a young adult audience. In this ask, I am focusing on ways to make the two main characters resonate more with this audience and will ask the model for suggestions on improvements that can be made to reach this goal.

```python
# Constructing the prompt
prompt_one = """
Context: You are a focus group reviewing film and television scripts and story synopses.
Review the following scene. Tell me how you think these characters would resonate with an audience of young adults, 
and provide suggestions on how to make the characters stronger.

Script:
EXT. DOG PARK - DAY
BEN and VERA sit on a park bench in a trendy dog park
watching BEN’s dog Lupin run around in circles.
VERA
You never told me how your date
went the other night.
BEN
It was … actually a little strange.
You know how Lupin is in a few of
my profile pictures, right? On the
app?
VERA
From what you’ve told me, Lupin is
an excellent wing-dog.
BEN
He’s the goodest boy there is.
VERA
So what happened?
BEN
So the other night, I’m out with
this guy … “Ted”. We had a great
time: drinks at that new place with
the secret door, dinner at La Mer
and then I suggested a night cap
back at mine. We get back to mine
and Ted meets Lupin and is all over
him--big cuddles, throwing toys
down the hall. Like I suddenly
didn’t exist.
VERA
Maybe you’re Lupin’s wingman, did
you ever think about that?
BEN
But Ted hadn’t mentioned him all
night! Most guys I’m out with,
that’s like the first thing they
bring up. Lupin’s a great
icebreaker, that’s why I’ve got him
in so many of my pictures.
(MORE)
BEN (CONT'D)
2.
This guy shows up at my place and
it’s like he just happens to
discover that I’m a dog owner?
VERA
Doesn’t really sound like much of a
crime, Ben.
BEN
Then: I go to the kitchen to get a
bottle of wine, and when I come
back Ted’s taking all these photos
with Lupin. Hugging him, patting
him, staring in his eyes. I swear I
caught him draping his scarf around
Lupin’s neck. I had to fake a
headache and get him out of there.
Like, that is weird, right?
Pause. VERA shifts her weight on the park bench.
VERA
Do you still have him on the app?
BEN forks over his phone. VERA scrolls for a moment and then
gasps.
VERA (CONT'D)
Oh. My. God. Look!
“Ted” has a different dog in every picture on his profile.
BEN swipes through them in horror, settling on one in which
Lupin wears a scarf. He looks over to the dog, oblivious,
running around the park with a stick in his mouth.
BEN
He’s a monster. Some kind of serialkiller-dog-stalker.
VERA
You’re so lucky. Both of you!
BEN
I need to hug Lupin. (Calling out.)
Lupin! LUPIN!
He runs to his dog across the grass.
"""
```

```python
# Calling the API
response = client.completions.create(
    model="gpt-3.5-turbo-instruct",
    prompt=prompt_one,
    max_tokens=500,
    temperature=0
)

# Getting the text response
response_text = response.choices[0].text.strip()

# Wrapping the text to a specified width
wrapped_text = textwrap.fill(response_text, width=80)

# Printing the formatted response
print(wrapped_text)
```

Overall, the model gave some fairly constructive feedback on both how the story would resonate with a young adult audience and how the characters could be improved. For the first part of my ask, the model gave some spcific ways that the story would resonate with young adults, citing how the theme of regaling a datng story, particularly one that involved dogs, would be relatable to this demographic. I was, however, hoping for a little more commentary on this ask, though this could easily be expanded on with some additional prompt engineering.

The model did provide a lot of fairly stong feedback for how the characters could be improved for this particular audience. Suggestions like adding more depth or individuality or providing more background on the characters to understand why this maay be a significant conversation for them are decent examples of feedback that can be used to either strengthen the scene or continue the story. However, the overally feedback does appear rather surface-level, indicating that some work will be needed to better specialize the model for this task.

### Prompt 2: Dialogue Tailoring

The next prompt I am going to experiment with will ask the model to take some preexisting dialogue and better tailor it to a different audience. This could be particularly useful if a writer needs to write for a particular demographic but might not be overly familiar with the nuances of that particular audience. 

```python
# Constructing the prompt
prompt_two = """
Context: You are a focus group reviewing film and television scripts and story synopses.
Review the following short script. Give suggestions on how to tailor the dialogue between the characters
more to a Gen Z audience. Give specific examples of where you would incorporate some of these suggestions in the script.

EXT. APARTMENT BALCONY - NIGHT
ASH checks their watch: just minutes left until midnight.
They look up and out at the city before them and sigh through
their nose.
They’re standing on the large, wraparound balcony of a nice
apartment. Inside, a New Year’s Eve party is in full swing.
Out here, in the cold, various guests have stepped out for
various reasons: for fresh air, for a smoke, for a break from
the crowd, for privacy with one another. ASH does their best
to ignore the lot of them--to keep their distance. They check
their watch again.
The sounds of the party crescendo for a moment as the sliding
balcony door is quickly opened and closed. Out of the party
steps ASH’s friend CHARLIE, carrying a bottle of champagne
and two glasses. CHARLIE joins ASH by the railing.
CHARLIE
(pouring drinks)
I’ve got it!
ASH
(taking a glass)
Nice! What have you got?
CHARLIE
I’ve been thinking.
ASH
You’ve been thinking...
CHARLIE
It’s a thought. A bold thought.
ASH
A New Year’s resolution?
CHARLIE
God, no! It’ll be far too late by
then.
ASH
So let’s hear it.
CHARLIE
Not yet: first, we drink.
They ‘cheers!’ And drink. CHARLIE pours another round.
2.
ASH
So far, so good.
CHARLIE
This isn’t even “the thing”.
ASH
So what is “the thing”?
CHARLIE
Drink again and I’ll tell you.
ASH drinks. CHARLIE doesn’t.
CHARLIE (CONT'D)
I think we should kiss at midnight.
Pause. ASH performs the longest drink swallow of their life.
CHARLIE fills ASH’s glass again.
ASH
You and me?
CHARLIE
Yep.
ASH
You want to kiss?
CHARLIE
At midnight. For New Year’s.
ASH
I’m confused.
CHARLIE
Don’t be. It’s a tradition, whereASH
No, I get that bit. You want us to
kiss at midnight?
CHARLIE
I figure with you, I’m guaranteed a
kiss. If I look elsewhere, I might
not be so lucky.
ASH
So it’s insurance? Isn’t that guy
here? That one you want to...
CHARLIE
Nathan. Yes, Nathan is here. And I
thought about shooting that shotASH
But instead2.
3.
CHARLIE
...I came out here.
ASH
For me?
CHARLIE
Little bit wishing I hadn’t, now...
An OBNOXIOUS GUEST bangs on the glass door and mouths “GET
READY!” Charlie downs their glass and pours another.
ASH
What are you doing?
CHARLIE
Catching up with you.
ASH
You know that’s not what I meant.
A “WHOOP!” from the crowd inside. CHARLIE smiles at ASH.
CHARLIE
It might be nice?
ASH
I think it could be.
CHARLIE
Not where I’d planned the evening
was gonna go...
ASH
You can still go and find Nathan.
CHARLIE shakes their head. Inside the apartment, a muffled
countdown begins: *TEN*, *NINE*, *EIGHT*, *SEVEN*...
CHARLIE
So are we doing it?
ASH
Is it a good idea?
CHARLIE
That sounds like next year’s
problem.
ASH
Just a kiss?
CHARLIE
Maybe?
... *THREE*, *TWO*, *ONE*! The party erupts. Champagne,
cheering, laughter, hugs, kisses. And ASH and CHARLIE
together in the middle of it all.
"""
```

```python
# Calling the API
response = client.completions.create(
    model="gpt-3.5-turbo-instruct",
    prompt=prompt_two,
    max_tokens=500,
    temperature=0
)

# Getting the text response
response_text = response.choices[0].text.strip()

# Wrapping the text to a specified width
wrapped_text = textwrap.fill(response_text, width=80)

# Printing the formatted response
print(wrapped_text)
```

This output was significantly less helpful than that from the first prompt experiment. The model leaned heavily into Gen Z stereotypes, to the point where if all of these suggestions were incorporated I guarantee Gen Z would not give this film the time of day. While it was good at pulling out things that are important to Gen Zers and provide suggestions on how to incorporate them, it lacked the ability to incorporate these ideas into the script in a way that would feel organic and natural. I also noticed that the model focused more on the stage direction parts of the scripts as opposed to the dialogue, indicating that some training may be needed to have the model better understand the structure of a script. 

### Prompt 3: Conceptual Brainstorming/Story Mapping

For the next prompt, I am taking a step away from direct script review and focusing more on overall story and narrative structure. I assign the model a similar role as previously and provide a short synopsis for a short film, asking for feedback on the plan outlined in the synopsis for character development and story pacing. The goal of this prompt is to see how good the model might be in the early stages of story developmet at providing ideas and thoughts to the writer to incorporate as they build out their story and screenplay.

```python
prompt_three = '''
Context: You are a focus group reviewing film and television scripts and story synopses.

Imagine a short film where the protagonist is a young artist struggling to find inspiration. The film follows the protagonist 
as they go about their day to day life seemingly in a creative slump. However, something happens that opens the artist's eyes
up to the wonder in the ordinary, and we watch as the artist turns their ordinary experiences into extraordinary art.

Describe some feedback you might give on the character development and story pacing.'''
```

```python
# Calling the API
response = client.completions.create(
    model="gpt-3.5-turbo-instruct",
    prompt=prompt_three,
    max_tokens=500,
    temperature=0
)

# Getting the text response
response_text = response.choices[0].text.strip()

# Wrapping the text to a specified width
wrapped_text = textwrap.fill(response_text, width=80)

# Printing the formatted response
print(wrapped_text)
```

The model produced some rather insightful feedback for this prompt. In this proposed story about an artist slowly finding their inspiration, the pacing and character development would be two of the most critical parts of the narrative. The model's suggestion to explore some of the artist's journey to become an artist would certainly lend well to this kind of short film, potentially making the impact of the resolution even more impactful to see the artist's journey from beginning to end. On the pacing front, the model's suggsetion to have plenty of examples of the artist's struggle and frustration with not finding inspiration is also sound advice to really set the story up to have a great emotional impact when the resolution came. While this is great general feedback, the model still struggles with providing specific ways in which some of these narrative hits can be achieved, indicating that more specific prompting or specialized training would be needed to really fill this role.
